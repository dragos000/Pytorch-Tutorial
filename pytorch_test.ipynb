### **Video #4**

```python
tensor_a = torch.tensor([1,2,3,4])
tensor_b = torch.tensor([5,6,7,8])
```

```python
# Addition
tensor_a+tensor_b
```

```python
# Addition Longhand
torch.add(tensor_a,tensor_b)
```

```python
# Subtraction
tensor_b-tensor_a
```

```python
torch.sub(tensor_b,tensor_a)
```

```python
# Multiplication
tensor_a*tensor_b
```

```python
torch.multiply(tensor_a,tensor_b)
```

```python
tensor_b/tensor_a
```

```python
torch.div(tensor_b,tensor_a)
```

```python
# Remainder Modulus
tensor_b % tensor_a
```

```python
torch.remainder(tensor_b,tensor_a)
```

```python
# Exponents / power
torch.pow(tensor_a,tensor_b)
```

```python
# Another way to use the long form
tensor_a.add(tensor_b)
```
